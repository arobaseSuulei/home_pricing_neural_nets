```python
X=[]
y=[]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_X = StandardScaler()
```

```python
from random import randint
#generer mon dataset
for _ in range(2000):
  surface=randint(50,200)
  chambre=randint(1,6)
  age=randint(0,20)

  prix=(1000 * surface
    + 5000 * chambre
    - 500 * age
    + 15 * surface * chambre)

  X.append([surface,chambre,age])
  y.append(prix)



```

```python
for i in range(30):
  print(X[i],y[i])
```

```python
# la on convertit en tableau numpy, keras travaille mieux avec des tabs numpy

import numpy as np

X=np.array(X)
y=np.array(y)

print(X.shape)
print(y.shape)
```

```python
# le shuffle

indices=np.random.permutation(len(X))

X=X[indices]
y=y[indices]
```

```python
# on split
X_train=(X[:1600])
X_test=(X[1600:])

y_train=(y[:1600])
y_test=(y[1600:])
```

```python
# @title Normalisation
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=tf.keras.Sequential()

#premiere couche
model.add(keras.layers.Dense(
    16,
    activation='relu',
    input_shape=(3,)
)
)
```

```python
#2e couche
model.add(keras.layers.Dense(
    16,
    activation='relu',

)
)
```

```python
model.add(keras.layers.Dense(1)) #sortie
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
history=model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32
)
```

```python
# @title Normalisation
maison=np.array([[108, 5, 1]])
maison=scaler_X.transform(maison)
prediction=model.predict(maison)

```

```python
prediction=model.predict(
    maison
)
```

```python
print(prediction)
```

```python
loss,mae=model.evaluate(X_test,y_test)
```
